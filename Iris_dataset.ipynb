### Import Libraries

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

### Import Dataset

```python
data = pd.read_csv('Iris.csv')
print(data)
```

### For this data set there are 4 features. Now we will separate the features from the target.

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
x = data.loc[:, features].values
print(x)
```

### Now we will separate the target

```python
y = data.loc[:,['species']].values
print(y)
```

### Now we will Standardize the feature. StandardScaler standardizes a feature by subtracting the mean and then scaling to unit variance.

### To know more about StandardScalar please check out the link: https://towardsdatascience.com/scale-standardize-or-normalize-with-scikit-learn-6ccc7d176a02

```python
x = StandardScaler().fit_transform(x)
print(x)
```

### Now we will do feature reduction using PCA. In our dataset we have 4 features and we will project them here in 2 features.

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
finalDf = pd.concat([principalDf, data[['species']]], axis = 1)
finalDf.head()
```

### Now we will visualize the 2D Projection

```python
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 13)
ax.set_ylabel('Principal Component 2', fontsize = 13)
ax.set_title('PCA on Iris Dataset for 2D Projection', fontsize = 20)
targets = ['setosa', 'versicolor', 'virginica']
colors = ['y', 'm', 'c']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['species'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

### From this visualization above we can infer that, Versicolor and Virginica are some how more similar in terms of features than Setosa. Setosa demonstatres more independent and unique feature.
